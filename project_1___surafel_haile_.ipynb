```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.style.use('ggplot')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Reben80/Data110-22016/f38fe2d7cad650dda1c96dfde6297a9a95b8d410/dataset/box_office_data-2.csv')
```

```python
data={'Movies':['Argylle'	,'Lisa Frankenstein','The Beekeeper','Wonka','Migration'],
      'Weekly gross':[6249360,3695785,3398372,3056430,2935590],
      'Theaters':[3605,3144,3057,2764,2684]}
dm = pd.DataFrame(data)
dm['Average'] = dm['Weekly gross']/dm['Theaters']
dm_sorted= dm.sort_values(by='Average',ascending=True)
color= ['g','k','b','r','m','y','c']
plt.barh(dm_sorted['Movies'],dm_sorted['Average'],color=color)
plt.title('Average Earnings (per Theater) for Top 5 movies')
plt.xlabel('Movies')
plt.ylabel('Average earnings')
plt.show()
```

Here, we see what the average earnings per theater is for each of the top 5 movies in the Box Office .

```python
mean_average_earnings = dm_sorted['Average'].mean()

dm_sorted['Percentage_Difference_from_Mean'] = ((dm_sorted['Average'] - mean_average_earnings) / mean_average_earnings) * 100


dm_sorted_percent = dm_sorted.sort_values(by='Percentage_Difference_from_Mean', ascending=True)


colors = ['red' if x < 0 else 'green' for x in dm_sorted_percent['Percentage_Difference_from_Mean']]

plt.figure(figsize=(10, 5))
plt.barh(dm_sorted_percent['Movies'], dm_sorted_percent['Percentage_Difference_from_Mean'], color=colors)
plt.xlabel('Percentage Difference from Overall Mean Average Earnings (%)')
plt.ylabel('Movies')
plt.title('Percentage Difference in Average Earnings (per Theater) from Overall Mean')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.axvline(0, color='grey', linestyle='--', linewidth=0.8) # Add a line at 0 for reference
plt.tight_layout()
plt.show()
```

The graph you're seeing shows how much each movie's average earnings per theater deviates from the overall average of all five movies.

Green bars indicate movies that performed better than the overall average in terms of earnings per theater. The longer the green bar, the higher their earnings were compared to the mean. For example, 'Argylle' is significantly above the mean, meaning it earned substantially more per theater than the average of all five movies.

Red bars indicate movies that performed worse than the overall average in terms of earnings per theater. The longer the red bar, the lower their earnings were compared to the mean. For instance, 'Migration' is considerably below the mean, suggesting it earned much less per theater than the average of the five movies.

In essence, it highlights which movies are pulling their weight (or more!) in terms of efficiency per theater, and which ones are lagging behind the group's average.

```python
df = pd.DataFrame({
    "movie": ["Argyle","Migration","The Beekeeper","Wonka"],
    "rank": [1,5,3,4],
    "lastweek_rank": [1,5,3,4]})

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df,x="rank",y="lastweek_rank",s=50,color='b',edgecolor="black")

max_rank = max(df["rank"].max(), df["lastweek_rank"].max())
line_values = np.arange(1, max_rank + 1)

plt.plot(line_values, line_values, linestyle="--", linewidth=1.5,color="gray",label="No Change")

plt.title("Movie Rank Changes")
plt.xlabel(" Rank ")
plt.ylabel("last week Rank")
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.legend()
plt.grid(True, linewidth=0.3)

plt.show()
```

```python
df = pd.DataFrame({
    "movie": ["I.S.S.","Scrambled","Fallen Leaves","Monster"],
    "rank": [31, 41, 50, 46,],
    "last week_rank": [18, 19, 62,55]})

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df,x="rank",y="last week_rank",s=50,color='b',edgecolor='black')

max_rank = max(df["rank"].max(), df["last week_rank"].max())
line_values = np.arange(1, max_rank + 1)

plt.plot(line_values, line_values, linestyle="--", linewidth=1.5,color="gray",label="No Change")

plt.title("Movie Rank Changes")
plt.xlabel("Rank")
plt.ylabel("Last week's Rank")
plt.gca().invert_xaxis()
plt.gca().invert_yaxis()
plt.legend()
plt.grid(True, linewidth=0.3)

plt.show()
```

These graphs show how well the top movies keep their ranks versus how well movies at the lower end of the data set do so . If a movie kept it's ranking, than we see it's stats on the 'No change' line. On the ohter hand the movies that are found on the bottom of the box office list seem to be losing their 'release rank'.
