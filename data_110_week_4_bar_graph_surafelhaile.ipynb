```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

Task 1:

```python
plt.style.use('ggplot')
```

```python
x=['A.A','Burglary','Larceny','M.V.T','Murder','Rape','Robbery']
y=[264.1,250.7,1347.2,318.7,5.7,38.0,66.5]


data = {'Crime': x, 'Frequency': y}
dm = pd.DataFrame(data)


dm_sorted = dm.sort_values(by='Frequency', ascending=False)

plt.title('Crime Rates in 2023')
plt.xlabel('Crime')
plt.ylabel('Frequency')
plt.bar(dm_sorted['Crime'], dm_sorted['Frequency'], color='orange')
plt.tight_layout()
plt.show()
```

Task 2:

```python
data={'Crimes':['A.A','Burglary','Larceny','M.V.T','Murder','Rape','Robbery'],
      'Frequency':[264.1,250.7,1347.2,318.7,5.7,38.0,66.5],
      'Frequency22':[260.8,242.6,1342.3,330.7,-6.3,28.2,65.7]}
dm = pd.DataFrame(data)
dm['Percentage Change'] = ((dm['Frequency'] - dm['Frequency22']) / dm['Frequency22']) * 100
dm_sorted= dm.sort_values(by='Percentage Change',ascending=False)
color= ['g','k','b','r','m','y','c']
plt.barh(dm_sorted['Crimes'],dm_sorted['Percentage Change'],color=color)
plt.title('Crime Rates (percentage change 2022-2023)')
plt.xlabel('Frequency')
plt.ylabel('Crime')
plt.show()
```


Why does this graph make more sense than the graph of task 1:

It is much more clearer for people to see crime as a flowing statistic rather than one that is exact year to year,for crime is constantly increasing and decreasing. The first graph is misleading for that reason.

```python

crimetype = ['A.A','Burglary','Larceny','M.V.T','Murder','Rape','Robbery']
values1 = [264.1,250.7,1347.2,318.7,5.7,38.0,66.5]
values2 = [260.8,242.6,1342.3,330.7,-6.3,28.2,65.7]

bar_width = 0.35
index = np.arange(len(crimetype))

plt.bar(index, values1, bar_width, label='Crime in 2023')
plt.bar(index + bar_width, values2, bar_width, label='Crime in 2022')

plt.xlabel('Type of Crime')
plt.ylabel('Frequency')
plt.title('Crime 2022 vs 2023')
plt.xticks(index + bar_width / 2, crimetype)
plt.legend()
plt.tight_layout()
plt.show()
```
