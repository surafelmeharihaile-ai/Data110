```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

Task 1:

```python
plt.style.use('ggplot')
```

```python
x=['A.A','Burglary','Larceny','M.V.T','Murder','Rape','Robbery']
y=[264.1,250.7,1347.2,318.7,5.7,38.0,66.5]
plt.title('Crime Rates in 2023')
plt.xlabel('Crime')
plt.ylabel('Frequency')
plt.bar(x,y,color='r')
plt.tight_layout()
plt.show()
```

Task 2:

```python
data={'Crimes':['A.A','Burglary','Larceny','M.V.T','Murder','Rape','Robbery'],
      'Frequency':[264.1,250.7,1347.2,318.7,5.7,38.0,66.5]}
dm = pd.DataFrame(data)
dm_sorted= dm.sort_values(by='Frequency',ascending=True)
color= ['g','k','b','r','m','y','c']
plt.barh(dm_sorted['Crimes'],dm_sorted['Frequency'],color=color)
plt.title('Crime Rates (percentage change)')
plt.xlabel('Frequency')
plt.ylabel('Crime')
plt.show()
```


Why does this graph make more sense than the graph of task 1:

It is much more clearer for people to see crime as a flowing statistic rather than one that is exact year to year. The first graph is misleading for that reason.
