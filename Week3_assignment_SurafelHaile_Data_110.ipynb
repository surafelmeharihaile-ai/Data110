**Part 1:** Practice

Simple Line Plots

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))
plt.show()
```

np.sin(x) and np,cos(x)

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

Line colors and styles.

SIN (X-0):

```python
plt.plot(x, np.sin(x-0),color="green",linestyle="dotted")
plt.show()
```

SIN (X+1):

```python
plt.plot(x, x + 0, linestyle='dashed', color='red')
plt.plot(x, x + 1, linestyle='dashdot', color='green')
plt.show()
```

Lines and colors combined:

```python
plt.plot(x, x + 0, '-b')
plt.plot(x, x + 1, '--p')
plt.plot(x, x + 2, '-.r')
plt.show()
```

Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))

plt.xlim(0, 11)
plt.ylim(-2.5, 2.5)
```

reverse limit edit :

```python
plt.plot(x, np.sin(x))

plt.xlim(11, 0)
plt.ylim(2.2, -2.2)
```

plt.axis:

```python
plt.plot(x, np.sin(x))
plt.axis("off")
```

```python
plt.plot(x, np.sin(x))
plt.axis('square')
```

Labeling Plots:

```python
plt.plot(x, np.cos(x),"--b" ,label="cos(x)")
plt.plot(x, np.sin(x),"-c" ,label="sin(x)")
plt.axis('square')

plt.legend()
plt.show()
```

Explain the difference between x as domain vs. as an index?

X as a domain represents a possible input value of the funtion np.sin(X) or cos (x). While x as a index represents the position of a data point within the graph.

```python
import numpy as np
import matplotlib.pyplot as plt
```

1: Setup and Styles

2:Basic line Plot

```python
x=np.linspace(-2,4,10)
y=x**2-2*x+1
plt.plot(x,y)
plt.show()
```

3: Multiple lines

```python
x=np.linspace(-2,4,10)
y1=x**3
y2=np.exp(x)
y3=np.log(x, where = x>0)
plt.plot(x,y1,label='fx')
plt.plot(x,y2,label='gx')
plt.plot(x,y3,label='hx')
plt.legend()
plt.show()
```

4:Styling lines

```python
x=np.linspace(-2,4)
y1=x**2
y2=x**3
y3=x**4
y4=x**5
plt.plot(x,y1,label='fx',color='r')
plt.plot(x,y2,label='gx',color=(1.0,0.2,0.3))
plt.plot(x,y3,label='hx',color='green')
plt.plot(x,y4,label='ix',color='b')
plt.legend()
plt.show()
```

5:Axes limits and Scaling

```python
x=np.linspace(-10,10)
y=1/x**2+1
plt.plot(x,y)
plt.axis('tight')
plt.xlim(-10,10)
plt.ylim(-35,35)
plt.show()
```

6: Labels and Legends

```python
x=np.linspace(-10,10,10)
y1=np.square(x)
y2=x**1/3
plt.plot(x,y1,label='fx')
plt.plot(x,y2,label='gx')
plt.legend(loc = "upper right")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

7: Basic scatter plot

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.plot(x,y,marker="o")
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
plt.show()
```

8: Customizing Scatter Plots

```python
x=np.linspace(0, 4*np.pi,)
y=np.sin(x)
plt.plot(x,y,marker="o",markersize=7,markerfacecolor='w',markeredgecolor='k')
plt.show()
```
