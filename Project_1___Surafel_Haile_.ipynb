```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
box_office=pd.read_csv('https://raw.githubusercontent.com/Reben80/Data110-22016/7f640882c31a40f469d392ff08d11e225f7e4f2d/dataset/box_office_data-2.csv')
box_office.head()
```

```python
sns.barplot(data=box_office, x='distributor',y='rank', errorbar=None)
plt.title('Distributor and Rank')
plt.xlabel('Distributor')
plt.ylabel('Rank')
plt.show()
```

```python
sns.barplot(data=box_office, x='',y='', errorbar=None)
plt.tight_layout()
plt.title('Movie and Year')
plt.xlabel('Distributor')
plt.ylabel('Year')
```
