```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.style.use('ggplot')
```

```python
data={'Crimes':['A.A','Burglary','Larceny','M.V.T','Murder','Rape','Robbery'],
      'Frequency':[264.1,250.7,1347.2,318.7,5.7,38.0,66.5],
      'Frequency22':[260.8,242.6,1342.3,330.7,-6.3,28.2,65.7]}
dm = pd.DataFrame(data)
dm['Percentage Change'] = ((dm['Frequency'] - dm['Frequency22']) / dm['Frequency22']) * 100
dm_sorted= dm.sort_values(by='Percentage Change',ascending=False)
color= ['g','k','b','r','m','y','c']
plt.barh(dm_sorted['Crimes'],dm_sorted['Percentage Change'],color=color)
plt.title('Crime Rates (percentage change 2022-2023)')
plt.xlabel('Frequency')
plt.ylabel('Crime')
plt.show()
```
