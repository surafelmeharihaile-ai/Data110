```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Reben80/Data110-22016/refs/heads/main/dataset/world_countries.csv')
df.head()
```

```python
df.groupby('Region')['GDP'].mean().plot(kind='bar')
plt.title('Average GDP by Region')
plt.xticks(rotation=75)
plt.xlabel('Region')
plt.ylabel('Average GDP')
plt.show()
```

```python
plt.scatter(df['GDP'], df['Infant mortality'])
plt.xlabel('GDP')
plt.ylabel('Infant Mortality')
```
